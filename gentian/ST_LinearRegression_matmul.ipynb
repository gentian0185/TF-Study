```python
import tensorflow as tf
import numpy as np
import random

#https://ko.wikipedia.org/wiki/%EB%82%9C%EC%88%98
#seed 라 불리는 수를 인자로 매번 다르게 주어 매번 다른 난수를 추출하게끔한다.
#그래서.. 지정하면 같은값이 나와야 하는데 계속 값이 변한다.
tf.set_random_seed(22)


# param
learning_rate = 0.001
training_epochs = 100
batch_size = 20


x_train = [[1,2,3],[4,5,6],[7,8,9]]
#x_train = np.zeros((3,3))
y_train = [1,2,3]

print(x_train)

#x_test = [4,5,6]
#y_test = [4,5,6]


X = tf.placeholder("float", [3, 3])
Y = tf.placeholder("float", [3])


# 안의 숫자는 shape
i = tf.random_normal([3,1]);
j = tf.random_normal([1]);
print(i, j)
W = tf.Variable(i, name='weight')
b = tf.Variable(j, name='bias')


# H(x) = Wx + b
hypothesis = tf.matmul(X, W) + b


# cost function
# square : 제곱
# reduce_mean : 평균을 내주는 함수.
cost = tf.reduce_mean(tf.square(hypothesis - Y))


# Cost minimaze..  0에 가깝도록 변경을 시켜주는 역할.
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
train = optimizer.minimize(cost)
# 또는 train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)


# init session
sess = tf.Session()
# tf.Variable 함수를 사용한 경우에는 반드시 사용해야함.
sess.run(tf.global_variables_initializer())


#start train
for epoch in range(training_epochs):
    total_batch = int(training_epochs/batch_size)
    
    for i in range(total_batch):
        cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X:x_train, Y:y_train})
        #cost_val, W_val, _ = sess.run([cost, W, train], feed_dict={X:x_train, Y:y_train})
        # (x) cost_val, W_val, _ = sess.run([cost, W, optimizer], feed_dict={X:x_train, Y:y_train})
        # sess.run(train, feed_dict={X:x_train, Y:y_train})
        # sess.run(train)

    #print('Epoch:', '%03d' % (epoch + 1), cost_val, W_val, b_val)
    #
    #print('Epoch:', '%03d' % (epoch + 1), sess.run(cost), sess.run(W), sess.run(b))

    
    
print('Learning Finished!')








# correct prediction test model
#prediction = tf.arg_max(hypothesis, 1)
#is_correct = tf.equal(prediction, tf.arg_max(Y, 1))
# 위 함수는 이것으로 변경될 수 있다.
# correct_prediction = tf.equal(tf.arg_max(hypothesis, 1), tf.arg_max(y_test, 1))
#accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

#print("Prediction:", sess.run(prediction, feed_dict={X: x_test}))
# Calculate the accuracy
#print("Accuracy: ", sess.run(accuracy, feed_dict={X: x_test, Y: y_test}))




    
```
